```python
compat_blood = {"O": {
    "O": 1,
    "A": 1,
    "B": 1,
    "AB": 1
},
"A": {
    "O": 0,
    "A": 1,
    "B": 0,
    "AB": 1
},
"B": {
    "O": 0,
    "A": 0,
    "B": 1,
    "AB": 1
},
"AB": {
    "O": 0,
    "A": 0,
    "B": 0,
    "AB": 1
}}
```

```python
data = """1 AB A
2 A AB
3 AB O
4 B O
5 A A
6 O B
7 A O
8 A O
9 A O
10 A AB
"""
```

```python
pairs = {}
l = 0
for line in data.strip().split("\n"):
    line = line.strip().split(" ")
    pairs[int(line[0])] = (line[2], line[1])

pairs
```

# Exercicio 1

```python
compat = {}

for i in range(1,len(pairs.keys())+1):
    for j in range(1,len(pairs.keys())+1):
        if i ==j:
            compat[(i,j)] = 0
        else:
            compat[(i,j)] = compat_blood[pairs[i][0]][pairs[j][1]]
```

```python
adj = {}

for i in range(1, len(pairs.keys()) + 1):
    for j in range(1, len(pairs.keys()) + 1):
        if compat[(i, j)] == 1:
            if i not in adj:
                adj[i] = []
            adj[i].append(j)
```

```python
adj
```

```python
cycles = get_all_cycles(adj,2)
for c in cycles:
    print(c)
print(len(cycles), "cycles", time.process_time(), "s")
z, sol = kep_cycle(adj,cycles)
print("solution:", z, sol)
```

```python
cycles = get_all_cycles(adj,3)
for c in cycles:
    print(c)
print(len(cycles), "cycles", time.process_time(), "s")
z, sol = kep_cycle(adj,cycles)
print("solution:", z, sol)
```

```python
import numpy as np

data_str = """
1 3 . 6 . 9 . 15 19 15 9
2 27 . 30 . . . . . . .
3 22 10 21 21 21 14 7 10 7 13
4 9 9 20 20 20 1 9 9 9 8
5 3 4 13 . 15 . 23 17 22 17
6 20 . 11 14 . . . . . .
7 1 . 10 11 10 . 10 11 8 4
8 11 . 10 10 10 5 . . . 2
9 14 . 4 5 3 5 19 13 18 25
10 23 . 34 . . . . . . .
"""

# Split into rows
rows = data_str.strip().split('\n')

# Parse each row
matrix = []
for row in rows:
    parts = row.split()
    # Skip the first element (patient or donor index)
    values = [float(x) if x != '.' else 0 for x in parts[1:]]
    matrix.append(values)

matrix
```

```python
len(matrix)
```

```python
life_exp = {}

for i in range(len(matrix)):
    for j in range(len(matrix)):
        if matrix[i][j] >5:
            if i+1 not in life_exp:
                life_exp[i+1] = []
            life_exp[i+1].append(j+1)

life_exp
```

```python


total_lf = {}

for i in range(1, len(adj)+1):
    total_lf[i] = list(set(adj[i]) & set(life_exp[i]))

```

```python
total_lf
```

```python
w = {}

for i in total_lf.keys():
    for j in total_lf.keys():
        w[(i,j)] = matrix[i-1][j-1]
```

```python
w
```

```python
cycles = get_all_cycles(total_lf, 3)

# Rodar a otimização
z, sol = kep_cycle_max_life_expectancy(total_lf, cycles, w)

print("Maximized life expectancy:", z)
print("Selected cycles:", sol)
```
